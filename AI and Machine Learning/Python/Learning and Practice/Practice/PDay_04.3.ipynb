```python
#Dynamic Printing Practice 
from IPython.display import display,Math
base = 3
exponent = 4
ans = base ** exponent
display(Math(('%g^%g = %g'%(base,exponent,ans))))
```

```python

#The Sigma Protocol 
exponent = 2
total_sum = 0 #
for base in range(2,8):
    ans = base ** exponent
    display(Math('%g^%g=%g'%(base,exponent,ans)))
    total_sum = total_sum + ans
    #Can Also be Written as total_sum += ans
    
print(f"The sum of squares from 2 to 7 is:{total_sum}")
```

```python
#The Boss (Polynomials )
x_values = [-2,-1,0,1,2]

for x in x_values:
    y = (2*(x**2))-(5*(x))+3
    display(Math('%g=2(%g)^2-5(%g)+3'%(y,x,x)))
    print(f"When X is {x}, Y is {y}")
```
