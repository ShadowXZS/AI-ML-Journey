```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#The Atom 
v = np.array([2,4])
w = np.array([3,1])

#The Operation 
z=v+w

#The Visualization 
plt.arrow(0,0, v[0], v[1], head_width=.05, width=0.02, color='red', label='v')
plt.arrow(v[0],v[1], w[0], w[1], head_width=.05, width=0.02, color='blue', label='w')
plt.arrow(0,0, z[0], z[1], head_width=.05, width=0.02, color='green', label='v+w')
plt.legend()
plt.axis([0,5,0,5])
plt.grid()
plt.show()
```
