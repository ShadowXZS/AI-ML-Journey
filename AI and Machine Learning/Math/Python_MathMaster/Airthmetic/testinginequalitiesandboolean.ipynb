```python
# Excercise 
from IPython.display import display,Math
x_value= [2 , 3, 4] # Presented with 3 Different Values of X, Hence Converted them to a List to apply for Loop
#1.

for x in x_value:
    ans = 4*x + 3 < 17 - x**2 
    display(Math('4\\cdot%g + 3 < 17 - %g^2 = %g '%(x,x,ans)))# \\cdot adds a . to Indicate Multiplication
    print(ans)

for x in x_value:
    ans2 = 8*x -2 <= -3*x +42
    display(Math('8\\cdot%g - 2 <= -3\\cdot%g +42 = %g'%(x,x,ans2)))# Need to be careful about %g syntax for LaTex
    print(ans2)
```
