```python
#Solution 
x = 7
y = -2 
z = 5
ans = 3*x*(4+y)
display(Math('3x(4+y) = %g' %ans))
display(Math('3\\times %g(4%g) = %g' %(x,y,ans)))
```

```python
ans = -y-(x+3)/z
display(Math('-y-\\frac{(x+3)}{z} = %g' %ans))
display(Math('-(%g)-\\frac{(%g+3)}{%g} = %g' %(y,x,z,ans)))
```

```python
# # # TakeAways 
# # ### ðŸ“ TAKEAWAYS: DYNAMIC MATH PRINTING

# # **1. Don't be a Human Calculator.**
# # * **Bad:** Calculating the result in your head and typing `42`.
# # * **Good:** `ans = 3*x*(4+y)` -> Let Python do the math.
# # * **Why:** If variables change, your hardcoded string becomes a lie.

# # **2. Variables over Values.**
# # * **Bad:** `Math('7 + 3')` (Hardcoded values).
# # * **Good:** `Math('%g + %g' %(x, y))` (Dynamic variables).
# # * **Rule:** The string should be a template; the variables provide the data.

# # **3. The Negative Sign Trap.**
# # * **Observation:** When `y = -2`, printing `-%g` results in `--2`.
# # * **Fix:** Use parentheses in your LaTeX template: `-(%g)` to make it look like `-(-2)`.

# # **4. The LaTeX/Python Bridge.**
# # * Use `\\frac{numerator}{denominator}` for fractions.
# # * Use `\\times` for multiplication (x).
# # * Use `%g` to keep numbers clean (removes `.0000`).

# ### ðŸ›‘ POST-MORTEM: HARDCODING VS. DYNAMIC PRINTING

# **The Error:**
# I initially "hardcoded" the numbers (calculated 42 in my head and typed `42`).
# * **Why it failed:** My code was a "Typewriter." If `x` changed to 10, my printout would still say 42 (a lie).
# * **The Goal:** My code must be a "Mirror." It should reflect whatever the variables hold.

# **The Fix (Dynamic Logic):**
# 1.  **Calculate First:** `ans = -y - (x+3)/z` (Let Python do the math).
# 2.  **Build Template:** Use `%g` as placeholders. `-(%g) ... = %g`.
# 3.  **Inject Variables:** Pass `(y, x, z, ans)` into the string.

# **The "Negative Sign" Trap:**
# * **Issue:** If `y = -2`, then `-%g` prints `--2`.
# * **Solution:** Use parentheses in the template: `-(%g)`. Result: `-(-2)`.

# **Key Syntax:**
# * `%g`: "General Format" (Removes ugly zeros like `5.0000` -> `5`).
# * `\frac{a}{b}`: LaTeX for Fractions.

#Writing Comments in Python 
#Anything After the # is not interpretred. 
#Organize your Code.
```
