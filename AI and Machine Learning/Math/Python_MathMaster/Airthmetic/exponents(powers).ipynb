```python
#Excercise 
#1. 
x = 5 
y = 5.1 
ans = x**(3/4)*4**y

display(Math('x^{3/4} \\times 4^y = %g'%ans))#Double Check where the Brackets are Placed.
display(Math('%g^{3/4} \\times 4^%g = %g'%(x,y,ans)))
```

```python
#2. 
x = 5 
y = 5.1 
ans = ((3**3)/(x**y))

display(Math('\\frac{3^3}{x^y} = %g'%ans))
display(Math('\\frac{3^3}{%g^%g} = %g'%(x,y,ans)))
```

```python
#3. 
x = 5 
y = 5.1 
ans = 10**(x-4)

display(Math('10^{x-4} = %g' %ans))
display(Math('10^{%g-4} = %g' %(x,ans)))
```
